## Import Library & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Telco Customer Churn.csv")
df.head(10)
```

```python
df.info()
```

```python
def viz_barplot(data, target_column, feature_column, x_axis_rotation = 0):
    """
    Data = dataframe
    Target_column = Churn (kolom tujuan)
    Feature_column = kolom lain yang ingin dilihat hubungannya dengan Churn
    """
    
    # fig = plt.figure(figsize = (7,5))
    ax = sns.countplot(data = data, x= feature_column, hue = target_column)
       
    ax.set_ylabel('Quantity')
    ax.set_xlabel(feature_column)
    plt.xticks(rotation = x_axis_rotation)
    plt.title(f'{feature_column} and {target_column}: Count & Proportion')
    plt.show()
```

```python
viz_barplot(data = df, target_column = 'Churn', feature_column = 'gender')
```

```python
viz_barplot(data = df, target_column = 'Churn', feature_column = 'Contract')
```

```python
viz_barplot(data = df, target_column = 'Churn', feature_column = 'Dependents')
```

```python
boxplot = sns.boxplot(data = df, x = "tenure", y = "Churn")
plt.show()
```
